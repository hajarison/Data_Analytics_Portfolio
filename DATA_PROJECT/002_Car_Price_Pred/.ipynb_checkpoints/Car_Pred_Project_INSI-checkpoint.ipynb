## Nettoyage du Dataset

```python
#Effacement des valeurs incomplete
data_clean= data.dropna()
```

```python
data_clean
```

```python
# Suppression du colonne inutile (Unnamed: 0)
data_clean=data_clean.drop('Unnamed: 0',axis=1)
data_clean
```

### Sur l'esnemble de données, il existe 3 type type de carburant : Petrol , Diesel , et le Compressed Natural Gas(CNG) 

### Problème : 
Pour les voiture qui on a de carburant PETROL ou DIESEL , Le kilometrage deja parcourrue sont mesurer en KILOMETRE PAR LITRE (kmpl)( sur le colonne qui s'appelle "Mileage" )
Le kilometrage du voiture de type de carburant CNG sont mesurer par KILOMETRE PAR KILOGRAMME (km/kg). 

J'ai décidé de suprrimer les voitures en CNG pour le moment car : 

    -  il y a juste 9 voitures qui ont ce type
    -  les valeur dans les tableaux sont en string donc il faut supprimer les unité avant de le convertir en variable de type FLOAT, Si je supprime tout les unité, cela signifie que tous les valeurs sont en même unité, c'est pas notre cas.
    -  il faut une conversion de kmpl en Km/Kg
    -   visualiser et explorer la majorité des données 



```python
#Extractions des voitures qui n'ont pas de type de carburant CNG
data_clean=data_clean[~data_clean['Fuel_Type'].str.contains('CNG')]
data_clean
```

## Effacement des unités et conversion des valeurs STRING en FLOAT ou INT

```python
# Puissance en bhp ou "brake horsepower" "Cheval-vapeur"(CV)
data_clean['Power(bhp)']=data_clean['Power'].str.replace(' bhp','').astype(float);
data_clean=data_clean.drop('Power',axis=1);
```

```python
# Moteur en CC ou "Cubic Centimeters" / "Centimetre Cube"
data_clean['Engine(CC)']=data_clean['Engine'].str.replace('CC','').astype(int)
data_clean=data_clean.drop('Engine',axis=1)
```

```python
# kilometrage en kmpl ou "Kilometers par litres" 
data_clean['km_driven(kmpl)']=data_clean['Mileage'].str.replace('kmpl','').astype(float)
data_clean=data_clean.drop('Mileage',axis=1) 
```

## Extraction des marques du voitures

```python
# Extraire le premier mot du colonne name pour obtenir les marque du voiture
data_clean['Marque']=data_clean['Name'].str.split(' ').str[0]
data_clean
```

```python
#Liste des marques 
data_clean['Marque'].unique()
```

```python
# Visualisation des marque les plus fréquents
plt.figure(figsize=(20,10))
sns.countplot(x='Marque',data=data_clean) 
```
