# Exercices de Visualisation de Données

## Les données

Cet ensemble de données classique contient les prix et autres attributs de près de 54 000 diamants. C'est un ensemble de données idéal pour les débutants qui apprennent à travailler avec l'analyse et la visualisation de données.

#### Contenu

* *price* : prix en dollars US (\$326 -- \$18,823)

* *carat* : poids du diamant (0,2 -- 5,01)

* *cut* : qualité de la coupe (Fair, Good, Very Good, Premium, Ideal) = (Moyen, Bon, Très bon, Supérieur, Idéal)

* *color* : couleur du diamant, de J (le pire) à D (le meilleur)

* *clarity* : mesure de la clarté du diamant (I1 (pire), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (meilleur))

* *x* : longueur en mm (0 -- 10,74)

* *y* : largeur en mm (0 -- 58,9)

* *z* : profondeur en mm (0 -- 31,8)

* *depth* : pourcentage total de la profondeur = z / moyenne (x, y) = 2 * z / (x + y) (43 -- 79)

* *table* : largeur du sommet du diamant par rapport au point le plus large (43 -- 95)

## Importations et Données

```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
diamonds = pd.read_csv('dataset/diamonds.csv')
```

## Tâches à effectuer

### Créer un diagramme de dispersion de la colonne prix par rapport à la colonne carat comme indiqué ci-dessous

```python
# Code ici
plt.figure(figsize =(6,4), dpi=72)
sns.scatterplot(x='carat', y='price', data=diamonds)
plt.show()
```

### Utiliser le paramètre alpha et le paramètre edgecolor pour traiter le problème de chevauchement et le problème de marqueur de bord blanc.

```python
# Code ici
plt.figure(figsize =(6,4), dpi=72)
sns.scatterplot(x='carat', y='price', data=diamonds, edgecolor=None, alpha=0.1)
plt.show()
```

### Agrandir le graphique précédent

```python
# Code ici
plt.figure(figsize =(12,8), dpi=72)
sns.scatterplot(x='carat', y='price', data=diamonds, edgecolor=None, alpha=0.1)
plt.show()
```

### Créer un histogramme de la colonne des prix comme indiqué ci-dessous

```python
# Code ici
plt.figure(figsize =(12,8), dpi=72)
myhist = sns.histplot(x='price', data=diamonds, edgecolor=None, alpha=0.4, bins=50)
myhist.set_xlim(0, 18000)
myhist.set_ylabel("")
plt.show()
```

### Créer un graphique de comptage des instances par type de coupe comme indiqué ci-dessous

```python
# Code ici
fig, ax = plt.subplots(figsize=(6, 4), dpi=72)
sns.countplot(data=diamonds, x='cut', ax=ax)
plt.show()
```

### Créer un grand diagramme en boîtes montrant la répartition des prix par type de coupe, comme indiqué ci-dessous

```python
# Code ici
plt.figure(figsize =(12,8), dpi=72)
sns.boxplot(x='cut', y='price', data=diamonds)
plt.show()
```

### Défi ! Voyez si vous pouvez modifier l'ordre des boîtes à moustaches comme indiqué ci-dessous. N'hésitez pas à changer également la coloration, pour refléter cette relation entre les types de coupe.

```python
# Code ici
plt.figure(figsize =(12,8), dpi=72)
sns.boxplot(x='cut', y='price', data=diamonds, order=['Fair', 'Ideal', 'Good', 'Very Good', 'Premium'], palette='blend:#45c5e4,#c545e4')
plt.show()

print(diamonds['cut'].unique())
```
