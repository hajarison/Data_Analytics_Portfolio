# EXERCICE SEABORN

```python
#importation des biblioth√®quqees
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('D:/Users/HAJARISON/INSI/Data/_DATA/dm_office_sales.csv')
df.head()
```

# SCATTER PLOT :

```python
#Nuage de point entre la colonne sales et salary
plt.figure(figsize =(15,9))
sns.scatterplot(x='salary', y='sales', data=df)
```

### NB :
#### hue :  regroupe les donn√©es en fonction des valeurs distinctes dans la colonne sp√©cifi√©e et attribue une couleur diff√©rente √† chaque groupe.

```python
plt.figure(figsize =(15,9))
sns.scatterplot(x='salary', y='sales', data=df, hue='division')
```

```python
plt.figure(figsize =(15,9))
sns.scatterplot(x='salary', y='sales', data=df, hue='work experience')
```


### NB :
#### palette :  Carte de couleurs

```python
plt.figure(figsize =(15,9))
sns.scatterplot(x='salary', y='sales', data=df, hue='work experience', palette='viridis')
```

### NB :
#### style :  style de points (forme : triangle, cercle ,...)

```python
plt.figure(figsize =(15,9))
sns.scatterplot(x='salary', y='sales', data=df, style='work experience')
```

### NB :
#### s : taille des style

```python
plt.figure(figsize =(15,9))
sns.scatterplot(x='salary', y='sales', data=df, style='work experience', hue='work experience', palette='viridis', s=100)
```

### Exportation de figure 

```python
plt.figure(figsize =(15,9))
sns.scatterplot(x='salary', y='sales', data=df, style='work experience', hue='work experience', palette='viridis', s=100)
plt.savefig("figure1.png")
```
