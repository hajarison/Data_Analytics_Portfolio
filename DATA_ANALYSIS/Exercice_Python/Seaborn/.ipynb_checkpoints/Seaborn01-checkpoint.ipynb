```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('C:\\Users\\ASUS_TMX\\Documents\\INSI L2\\Python\\Pandas\\_DATA\\dm_office_sales.csv')
df.head()
```

# SCATTER PLOT :

```python
plt.figure(figsize =(12,8))
sns.scatterplot(x='salary', y='sales', data=df)
```

### paramètre -> hue :
basé sur les features par rapport aux caractéristiques dans le dataframe

```python
plt.figure(figsize =(12,8))
sns.scatterplot(x='salary', y='sales', data=df, hue='division')
```

```python
plt.figure(figsize =(12,8))
sns.scatterplot(x='salary', y='sales', data=df, hue='work experience')
```

### paramètre -> palette :
Carte de couleurs (cmap)

```python
plt.figure(figsize =(12,8))
sns.scatterplot(x='salary', y='sales', data=df, hue='work experience', palette='viridis')
```

### paramètre -> style 
style de points

```python
plt.figure(figsize =(12,8))
sns.scatterplot(x='salary', y='sales', data=df, style='work experience')
```

### paramètre -> s :
s : size

```python
plt.figure(figsize =(12,8))
sns.scatterplot(x='salary', y='sales', data=df, style='work experience', hue='work experience', palette='viridis', s=100)
```

### export figure

```python
plt.figure(figsize =(12,8))
sns.scatterplot(x='salary', y='sales', data=df, style='work experience', hue='work experience', palette='viridis', s=100)
plt.savefig("plot.png")
```
